```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.dpi'] = 300

''' Tags
{'images': [], 'audio': [], 'histograms': [], 
'scalars': ['Train_EnvstepsSoFar', 'TimeSinceStart', 
'Train_AverageReturn', 'Training_Loss', 'Train_BestReturn'], 
'distributions': [], 'tensors': [], 'graph': False, 
'meta_graph': False, 'run_metadata': []}
'''

def load_event_return(folder, key='Train_AverageReturn'):
    ea = EventAccumulator(f'data/{folder}')
    ea.Reload()
    df = pd.DataFrame(ea.Scalars(key))
    return np.array(df['step']), np.array(df['value'])

def plot_experiments(title, experiments, x_label='Iterations', y_label='Train_AverageReturn'):
    plt.figure()
    for name in experiments:
        folder = experiments[name]
        steps, value = load_event_return(folder, y_label)
        plt.plot(steps, value)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, axis='y', linestyle='--')
    plt.legend(experiments.keys(), loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.show()
    plt.close()

def plot_curves(x, y_list, legends, title, x_label='Iterations', y_label='Train_AverageReturn'):
    plt.figure()
    for y in y_list:
        plt.scatter(x, y)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True, axis='y', linestyle='--')
    plt.legend(legends, loc='center left', bbox_to_anchor=(1, 0.5), fancybox=True, shadow=True)
    plt.show()
    plt.close()
```

```python
step, train = load_event_return('hw4_q2_obstacles_singleiteration_obstacles-cs285-v0_28-10-2022_20-02-37')
step, evalu = load_event_return('hw4_q2_obstacles_singleiteration_obstacles-cs285-v0_28-10-2022_20-02-37', 'Eval_AverageReturn')
plot_curves(step, [train, evalu], ['Train', 'Eval'], 'Q2', y_label='AverageReturn')
```
